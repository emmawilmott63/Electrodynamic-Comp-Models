```python
import numpy as np

# constants
c = 1
q = 1
m = 1

# functions
def magnitude(vector):
  return np.sqrt(np.sum(np.square(vector)))
def gamma_v(v):
  return 1/np.sqrt(1-magnitude(v/c)**2)
def gamma_u(u):
  return np.sqrt(1+magnitude(u/c)**2)

# boris algorithm
def boris(xn,un):
  C = q*dt/(2*m)
  gamman = gamma_u(un)
  xn_plus_half = xn + (un/(2*gamman))*dt
  uminus = un + C*E(xn_plus_half)
  gammaminus = gamma_u(uminus)
  T = (q*dt/(2*m*gammaminus))*B(xn_plus_half)
  s = 2*T/(1+magnitude(T))
  uprime = uminus + np.cross(uminus, T)
  uplus = uminus + np.cross(uprime, s)
  un_plus_one = uplus + C*E(xn_plus_half)
  gamma_plus_one = gamma_u(un_plus_one)
  xn_plus_one = xn_plus_half + (un_plus_one/(2*gamma_plus_one))*dt
  return xn_plus_one, un_plus_one, gamma_plus_one
```
